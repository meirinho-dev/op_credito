####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) 

import requests
import pandas as pd

def consulta_tesouro(nome_UF, estado_ou_municipio):
    # Montar a URL com base nos parâmetros fornecidos
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={nome_UF}&tipo_interessado={estado_ou_municipio}'
    
    # Fazer a requisição à API
    r = requests.get(url)
    
    # Converter a resposta em JSON
    data_json = r.json()
    print(data_json['items'])

# Verificar se a chave 'items' existe e tem dados
    if 'items' in data_json and data_json['items']:
        # Criar um DataFrame a partir dos dados
        
        participacoes = pd.DataFrame(data_json['items'])
        return participacoes
    else:
        print("Nenhum dado encontrado para os parâmetros fornecidos.")
        return pd.DataFrame()  # Retorna um DataFrame vazio
````

````python
# 2) 

# Definindo o UF de interesse
nome_UF = 'MG'  
# Definindo o tipo de interessado, como 'Estado' ou 'Município'
estado_ou_municipio = 'Município' 

# Chamar a função e obter os dados 
df_participacoes = consulta_tesouro(nome_UF, estado_ou_municipio)

# Exibindo os dados
print(df_participacoes.head())

````

````python
# Visualizando os dados em DataFrame
pd.DataFrame(df_participacoes.head())
````

````python
#Filtrando o status das solicitações 

pd.DataFrame(df_participacoes)['status'].value_counts()
````

````python
# 3) 

# Definindo o UF de interesse
nome_UF = 'BA'  

# Definindo o tipo de interessado, como 'Estado' ou 'Município'
estado_ou_municipio = 'Município'  

# Chamando a função para obter os dados 
df_participacoes = consulta_tesouro(nome_UF, estado_ou_municipio)

# Exibindo os dados importados
print(df_participacoes.head())
````

````python
# Visualizando os dados em um DataFrame
pd.DataFrame(df_participacoes.head())
````

````python
# Consultando a quantidade de solicitações por Município

pd.DataFrame(df_participacoes)['interessado'].value_counts()
````

````python
# 4) 

#Definindo o UF de interesse
nome_UF = 'BA'  

# Definindo o tipo de interessado, como 'Estado' ou 'Município'
estado_ou_municipio = 'Estado'  

# Chamando a função e obter os dados 
df_participacoes = consulta_tesouro(nome_UF, estado_ou_municipio)

# Exibindo os dados importados
print(df_participacoes.head())

````
